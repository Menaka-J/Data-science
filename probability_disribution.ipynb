```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

lambda_val = 4

probs = [poisson.pmf(i, mu=lambda_val) for i in range(15)]
plt.stem(range(15), probs)
plt.title(f'Poisson Distribution (lambda = {lambda_val})')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

p = 0.7

probs = [bernoulli.pmf(i, p) for i in [0, 1]]
plt.stem([0, 1], probs)
plt.title(f'Bernoulli Distribution (p = {p})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

mean = 0
std = 1

x_vals = np.linspace(mean - 4*std, mean + 4*std, 500)
probs = norm.pdf(x_vals, loc=mean, scale=std)

plt.plot(x_vals, probs)
plt.title(f'Normal Distribution (mean = {mean}, sd = {std})')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
import numpy as np

rate = 1  # rate = 1/lambda
x_vals = np.linspace(0, 5, 500)
probs = expon.pdf(x_vals, scale=1/rate)

plt.plot(x_vals, probs)
plt.title(f'Exponential Distribution (rate = {rate})')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
import pandas as pd
from scipy.stats import norm

df = pd.read_csv(r"C:\Users\Menaka\Music\datasets\2019 happy.csv")
for column in df.select_dtypes(['float', 'int']).columns:
    data_vals = df[column]
    mean, sd = data_vals.mean(), data_vals.std()
    x_vals = np.linspace(mean - 4*sd, mean + 4*sd, 500)

    plt.hist(data_vals, density=True, color='lightgrey', edgecolor='black', alpha=0.7)
    plt.plot(x_vals, norm.pdf(x_vals, loc=mean, scale=sd), color='darkblue', label='Normal Distribution')

    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
