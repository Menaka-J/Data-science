```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sb  
 
df = pd.read_csv(r"C:\Users\Menaka\Music\datasets\gym_members_exercise_tracking.csv") 
df.head() 
```

```python
import stemgraphic as sg  

total = df['Weight (kg)']  
sg.stem_graphic(total) 
```

```python
plt.figure(figsize = (15,8))  
plt.stem(df['Session_Duration (hours)'].values, df['BMI'].values)  
plt.title("Stem and Leaf plot for Session_Duration (hours) VS BMI")  
plt.xlabel("Session_Duration (hours)") 
plt.ylabel("BMI")  
plt.legend()  
plt.show() 
```

```python
plt.figure(figsize = (15,8))  
plt.stem(df['Session_Duration (hours)'].values, df['Fat_Percentage'].values)  
plt.title("Stem and Leaf plot for Session_Duration (hours) VS Fat_Percentage")  
plt.xlabel("Session_Duration (hours)")  
plt.ylabel("Fat_Percentage")  
plt.show() 
```
