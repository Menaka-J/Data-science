```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv(r"C:\Users\Menaka\Downloads\iris.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
sb.pairplot(data = df, hue = 'variety')
```

```python
sb.scatterplot(data = df, x = 'sepal.length', y = 'sepal.width', hue = 'variety') 
plt.title('Sepal length VS Sepal width')
```

```python
sb.lineplot(data = df, x = 'petal.length', y = 'petal.width', hue = 'variety') 
plt.title('Petal length VS Petal width')
```

```python
sb.barplot(data = df, x = 'variety', y = 'petal.length', hue = 'variety') 
plt.title('Count of Iris varieties using barplot')
```

```python
sb.countplot(data = df, x = 'variety', hue = 'variety')
plt.title('Count of Iris varieties using countplot')
```

```python
plt.figure(figsize = (15,8)) 
sb.barplot(data = df, x = 'petal.length', y = 'petal.width', hue = 'variety' ) 
plt.title('Petal length Vs Petal width')
```

```python
df_melt = pd.melt(df,id_vars=['variety'],value_vars=['sepal.length','sepal.width','petal.length','petal.width'])  
plt.figure(figsize=(10,7))  
sb.boxplot(x='variable',y='value',hue='variety',data=df_melt)  
plt.title("Boxplot of Iris Features by Variety")  
plt.show() 
```

```python
sb.displot(data = df, x = 'sepal.length', hue = 'variety') 
plt.title('Displot for sepal length ')
```

```python
sb.displot(data = df, x = 'petal.length', hue = 'variety') 
plt.title('Displot for petal length ')
```

```python
sb.histplot(data = df,  x = 'variety', hue = 'variety') 
plt.title('Histogram of Iris variety')
```

```python
corr = df.drop('variety', axis = 1).corr() 
print("CORRELATION MATRIX: \n") 
corr 
```

```python
plt.figure(figsize = (10,8)) 
sb.heatmap(corr, annot = True) 
plt.title("Heatmap of IRIS dataset")
```

```python
df.groupby('variety').agg(['mean', 'median'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))

sb.boxplot( y='petal.width', x= 'variety', data=df, orient='v' , ax=axes[0, 0], hue = 'variety')
sb.boxplot( y='petal.length', x= 'variety', data=df, orient='v' , ax=axes[0, 1], hue = 'variety')
sb.boxplot( y='sepal.length', x= 'variety', data=df, orient='v' , ax=axes[1, 0], hue = 'variety')
sb.boxplot( y='sepal.width', x= 'variety', data=df, orient='v' , ax=axes[1, 1], hue = 'variety')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))

sb.violinplot( y='petal.width', x= 'variety', data=df, orient='v' ,hue = 'variety', ax=axes[0, 0],inner='quartile')
sb.violinplot( y='petal.length', x= 'variety', data=df, orient='v' , ax=axes[0, 1], hue = 'variety',inner='quartile')
sb.violinplot( y='sepal.length', x= 'variety', data=df, orient='v' , ax=axes[1, 0], hue = 'variety',inner='quartile')
sb.violinplot( y='sepal.width', x= 'variety', data=df, orient='v' , ax=axes[1, 1], hue = 'variety',inner='quartile')
plt.show()
```
