```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Menaka\Downloads\iris.csv")
df.head()
```

RANDOM SAMPLING

```python
# Random Sampling
import numpy as np

# Set random seed for reproducibility (optional)
np.random.seed(42)

# Choose 1/3 of the data randomly to match the frequency of your systematic sample
sample_size = len(df) // 3
rand_index = np.random.choice(df.index, size=sample_size, replace=False)

# Create the random sample
randsample = df.loc[rand_index]
mean_rand = round(randsample['petal.width'].mean(), 3)

print("Mean of Random sample:", mean_rand)
print(randsample)
```

SYSTEMATIC SAMPLING

```python
index = np.arange(0,len(df), step=3)
syssample = df.iloc[index]
mean_sys = round(syssample['petal.width'].mean(), 3)
print("Mean of Systematic sample",mean_sys)
print(syssample)
```

CLUSTER SAMPLING

```python
n = 3

df['cluster_id'] = np.repeat([range(1,n+1)], len(df)/n)
index = []

#for this formula , cluster id must be even number
for i in range(0, len(df)):
    if df['cluster_id'].iloc[i]%2 == 0:
        index.append(i)

clustersample = df.iloc[index]
mean_cluster = round(clustersample['petal.width'].mean(), 3)
print("Mean of Cluster sample",mean_cluster)
print(clustersample)
```

STRATIFIED RANDOM SAMPLING 

```python
from sklearn.model_selection import StratifiedShuffleSplit as sss

#set the split criteria
split = sss(n_splits = 3, test_size = 8)

#perform data frame split
for x,y in split.split(df,df['variety']):
    stratifiedRsample = df.iloc[y].sort_values(by = 'sepal.width')

print(stratifiedRsample)
```

STRATIFIED RANDOM SAMPLING - ADDING SUBGROUPS

```python
df['strata'] = np.repeat([1,2,3], len(df)/3)
index = []

for i in range(0,len(df)):
    index.append(i)

stratRsample = df.iloc[index]
print(stratRsample)
```
